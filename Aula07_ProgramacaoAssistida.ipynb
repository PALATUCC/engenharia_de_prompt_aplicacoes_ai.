## Eh Primo

```python
def eh_primo(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

# Teste
numero = int(input("Digite um número: "))

if eh_primo(numero):
    print("É primo")
else:
    print("Não é primo")
```

## Inversão de string

```python
def inverte_texto(texto):
  invertido = ''
  for letra in texto:
    invertido = letra + invertido
  return invertido
```

## Mini Projeto

```python
# Calculadora
a1 = float(input("Digite um valor: "))
b = input("Digite sinal (+) ou (-): ")
a2 = float(input("Digite segundo valor: "))
if b == "+":
  resultado = a1 + a2
  print("Resultado: ", resultado)
elif b == "-":
  resultado = a1 - a2
  print("Resultado: ", resultado)
else:
  print("Operação invalida, coloque como sinal: (+) ou (-)")
```

```python
a = float(input("Digite primeiro Nº: "))
sinal = input("Digite o sinal: ")
b = float(input("Digite segundo Nº: "))

if sinal == "somar":
  resultado = a + b
  print("Resultado: ", resultado)
elif sinal == "subtrair":
  resultado = a - b
  print("Resultado: ", resultado)
else:
  print("Operação invalida, bote + ou - ")
```

```python
num1 = float(input("Digite o primeiro número: "))
operacao = input("Digite a operação (+), (-), (*) ou (/): ")
num2 = float(input("Digite o segundo número: "))

if operacao == "somar":
    resultado = num1 + num2
    print("Resultado:", resultado)

elif operacao == "subtrair":
    resultado = num1 - num2
    print("Resultado:", resultado)

elif operacao == "multiplicação":
    resultado = num1 * num2
    print("Resultado:", resultado)

elif operacao == "divisão":
    if num2 != 0:
        resultado = num1 / num2
        print("Resultado:", resultado)
    else:
        print("Erro: divisão por zero não é permitida")

else:
    print("Operação inválida")
```
