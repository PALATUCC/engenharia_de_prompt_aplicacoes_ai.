```python
# Script para verificar a temperatura do sistema

try:
    # Início de um bloco try-except para lidar com possíveis erros de entrada.
    # Simulando a obtenção da temperatura. Em um sistema real, você obteria isso de um sensor ou API.
    temperatura_str = input('Digite a temperatura atual do sistema em Celsius: ') # Solicita a temperatura ao usuário como string
    temperatura = float(temperatura_str) # Converte a string de temperatura para um número decimal (float)

    # Verifica se a temperatura é maior que 30 graus Celsius
    if temperatura > 30:
        # Se for maior, exibe um alerta de superaquecimento
        print('ALERTA: Superaquecimento! Temperatura atual: ' + str(temperatura) + '°C. Recomenda-se verificar o sistema.')
    else:
        # Se não for maior que 30, a temperatura é considerada normal
        print('Temperatura do sistema: ' + str(temperatura) + '°C. Tudo parece normal.')
except ValueError:
    # Se ocorrer um erro ao converter a entrada para float (ex: usuário digitou texto), exibe uma mensagem de erro
    print('Erro: Por favor, digite um número válido para a temperatura.')
```

## Missão 1 - Automação de arquivo

```python
import os
import shutil

# 1. Definir o diretório de teste
# O diretório onde os arquivos de exemplo serão criados e organizados.
# Se já existir, garantimos que ele esteja limpo para o teste.
test_directory = "arquivos_baguncados"

# Tenta remover o diretório e seu conteúdo se já existir, para um teste limpo.
if os.path.exists(test_directory):
    shutil.rmtree(test_directory)
# Cria o diretório de teste.
os.makedirs(test_directory)

print("Diretório de teste '{}' criado.".format(test_directory))

# 2. Criar 10 arquivos de exemplo
# Lista de nomes de arquivos para serem criados.
example_files = [
    "documento_importante.pdf",
    "foto_ferias.jpg",
    "relatorio_final.docx",
    "musica_favorita.mp3",
    "video_engracado.mp4",
    "nota_simples.txt",
    "apresentacao.pptx",
    "planilha_dados.xlsx",
    "backup.zip",
    "script_python.py",
    "outra_foto.png",
    "arquivo_log.log"
]

# Cria os arquivos vazios no diretório de teste.
for filename in example_files:
    file_path = os.path.join(test_directory, filename)
    with open(file_path, "w") as f:
        f.write("Conteúdo de exemplo para {}".format(filename))
    print("Arquivo '{}' criado.".format(filename))

print("\n--- Iniciando a organização dos arquivos ---")

# Dicionário para mapear extensões a nomes de pastas mais descritivos (opcional)
# Caso não esteja aqui, a pasta será criada com o nome da extensão (ex: "pdf", "jpg")
folder_names = {
    ".pdf": "PDFs",
    ".jpg": "Imagens",
    ".jpeg": "Imagens",
    ".png": "Imagens",
    ".gif": "Imagens",
    ".txt": "Textos",
    ".docx": "Documentos",
    ".xlsx": "Planilhas",
    ".pptx": "Apresentações",
    ".mp3": "Músicas",
    ".mp4": "Vídeos",
    ".avi": "Vídeos",
    ".mov": "Vídeos",
    ".zip": "Compactados",
    ".rar": "Compactados",
    ".py": "Scripts Python",
    ".log": "Logs"
}

# Percorrer todos os itens no diretório
for item in os.listdir(test_directory):
    # Construir o caminho completo para o item
    item_path = os.path.join(test_directory, item)

    # Verificar se é um arquivo (e não uma subpasta que já pode ter sido criada)
    if os.path.isfile(item_path):
        # Separar o nome do arquivo da extensão
        _, extension = os.path.splitext(item)
        # Converter a extensão para minúsculas para padronização
        extension = extension.lower()

        # Determinar o nome da subpasta
        # Se a extensão estiver no dicionário, usa o nome descritivo; senão, usa a própria extensão (sem o ponto)
        if extension in folder_names:
            subfolder_name = folder_names[extension]
        else:
            subfolder_name = extension[1:] + " (Outros)" # Mantendo a concatenação simples

        # Construir o caminho completo para a subpasta
        subfolder_path = os.path.join(test_directory, subfolder_name)

        # Criar a subpasta se ela não existir
        if not os.path.exists(subfolder_path):
            os.makedirs(subfolder_path)
            print("Pasta '{}' criada.".format(subfolder_name))

        # Construir o caminho de destino do arquivo
        destination_path = os.path.join(subfolder_path, item)

        # Mover o arquivo para a subpasta
        shutil.move(item_path, destination_path)
        print("Arquivo '{}' movido para '{}'.".format(item, subfolder_name))

print("\nOrganização concluída!")
print("Verifique o diretório '{}' para ver os resultados.".format(test_directory))
```
